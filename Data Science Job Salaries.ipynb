#  Preprocessing Data

```python
le=preprocessing.LabelEncoder()
lab=le.fit_transform(df['company_location'])
df['company_location']=lab
df
```

```python
le=preprocessing.LabelEncoder()
lab=le.fit_transform(df['company_size'])
df['company_size']=lab
df
```

```python
le=preprocessing.LabelEncoder()
lab=le.fit_transform(df['experience_level'])
df['experience_level']=lab
df
```

```python
le=preprocessing.LabelEncoder()
lab=le.fit_transform(df['employment_type'])
df['employment_type']=lab
df
```

```python
le=preprocessing.LabelEncoder()
lab=le.fit_transform(df['job_title'])
df['job_title']=lab
df
```
